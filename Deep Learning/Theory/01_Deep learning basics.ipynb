# 1. Universal Approximation Theorem (UAT)

Statement (informal)
- A feedforward network with a single hidden layer and a nonlinear, bounded, continuous activation can approximate any continuous function on a compact domain to arbitrary accuracy.

- Capacity result: existence of parameters $(\alpha_jg, w_j, b_j)$ for desired accuracy.
- Not a training guarantee; depth is not required by the theorem, but often improves efficiency.

## UAT Intuition: Building with Simple Bumps
- Each activated neuron acts like a soft "step" or "bump".
- Weighted sums of bumps form peicewise-smooth curves.
- Increasing hidden units refines approximation granularity.

## Key Takeaways
- Linear regression captures only hyperplanes; inadequate for general nonlinear structure.
- Neural networks = linear maps + nonlinear activations => flexible function families.
- UAT: a single hidden layer has enough capacity to approximate any continuous function on a bounded set.
- Depth is about efficiency and inductive bias: fewer units, better compositional structure.
- Theorem guarantees representability, not learnability or generalization.

# Forward pass and Loss Functions

## The forward pass: From Input to Prediction
The forward pass is the process of taking an input vector x and feeding it through the network, layer by layer, to produce a final prediction $\hat{y}$.
- Information flows sequentially through linear layers and non-linear activations.

The computation for every layer follows the same two-step, vectorized pattern:
- Linear Combination: Calculate the weighted sum for all the neurons in the layer. $$z^{[l]} = W^{[l]}a^{[l-1]} + b^{[l]}$$
- Non-linear Activation: Apply the activation function element-wise
$$a^{[l]} = g(z^{[l]})$$
The final activation vector $a^{[L]}$ is the network's prediction $\hat{y}$

## Loss Functions: Quantifying the Error
A loss function takes the true label (y) and the prediction $(\hat{y}$ and returns a single number representing the error. The error is what we aim to minimize during training.

### For Regression
MSE
- Calculates the average squared difference between true and predicted values.
- Penalizes large errors very heavily

### For Classification
Cross Entropy Loss
- $y_c$ is 1 if c is the true class, 0 otherwise
- $\hat{y_c}$ is the predicted probability for class c
- Penalizes the model for being confident and wrong.

Example (3 classes):
- True: Cat -> [1,0,0]
- Good Pred: [0.8,0.1,0.1] -> Low Loss
- Bad Pred: [0.1,0.2,0.7] -> High Loss

# Gradient Descent
## The Intuition
Imagine you are on a mountainside in thick fog and want to get to the lowest valley. What do you do?

You ask two simple questions at every step:

1. Which way is downhill?
- You feel the slope under your feet
- The direction of steepest ascent is the Gradient
- So, you take a step in the opposite direction

2. How big a step should I take?
- This is your "step size" or Learning Rate.
- Too small: takes forever. Too big: overshoots the valley.

## Why Gradient Descent? Navigating the Loss Landscape
The Goal
- Find the parameters $\theta$ that minimize the loss function $L(\theta)$. This means finding the lowest point in the landscape.

## GD in Action: Linear Regression

1. The Model
- Find the best line $\hat{y} = mx+c$. Parameters are $\theta = {m,c}$

2. The Loss Function (MSE)
$$L(m,c) = \frac{1}{N}\Sigma(\hat{y_i}-y_i)^2$$

3. The Update Rules
$$m = m-\eta\frac{\delta L}{\delta m}$$
$$c = c - \eta\frac{\delta L}{\delta c}$$
